```python
def generate_kmers_list(kmerlen): # this is from 2013 paper.
    nts = ['A', 'C', 'G', 'T']
    kmers = []
    kmers.append('')
    l = 0
    while l < kmerlen:
        imers = []
        for imer in kmers:
            for nt in nts:
                imers.append(imer+nt)
        kmers = imers
        l += 1

    return kmers
```

```python
import numpy as np
import os
```

```python
kmers=generate_kmers_list(6)
```

```python
def type_to_prob_matrix(MOTIF):
    iupac_abbr={'A':'A','C':'C','G':'G','T':'T','R':'AG','Y':'CT','S':'GC','W':'AT','K':'GT','M':'AC','B':'CGT','D':'AGT','H':'ACT','V':'ACG','N':'ATGC','-':''}
    atgc_d=dict(zip(['A','C','G','T'],[0,1,2,3]))
    A=np.zeros((len(MOTIF),4))#4列モチーフ長の行列を用意し、
    i=0
    for n in MOTIF:# type in した　motifを一文字ずつ分解して、
        dna=iupac_abbr[n]# IUPACコードを分解してAGとかATとかに変換して
        for d in dna:#その変換されたAGとかについて、AのポジションとGのポジションに1を入れる
            A[i][atgc_d[d]] +=1
        i +=1
    return(A)
```

```python
MOTIFS=kmers
names=kmers
meme_header='MEME version 4\nALPHABET= ACGT\nBackground letter frequencies\nA 3.197e-01 C 1.803e-01 G 1.803e-01 T 3.197e-01\nMOTIF '
al=4
mn=dict(zip(MOTIFS,names))
for MOTIF in mn:
    lg=len(MOTIF)
    A=type_to_prob_matrix(MOTIF)
    prm_header=" ".join(['\nletter-probability matrix: alength=',str(al),' w=',str(lg)])
    with open(os.path.join('query_6-mer_PM',mn[MOTIF]), mode='a') as outfile:
        outfile.writelines((meme_header,mn[MOTIF],prm_header,'\n'))
        np.savetxt(outfile,(A.T/sum(A.T)).T,fmt="%.5f")
```
